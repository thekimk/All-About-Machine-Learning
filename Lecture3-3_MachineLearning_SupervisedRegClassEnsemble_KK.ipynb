> **"`Train` 데이터의 `Bias가 적절(낮게)`한지 확인 후, `Test 데이터`에 적용하여 `Variance가 적절(낮게)하도록 반복적 업데이트`"**
>
> - Train의 Bias가 높다면,  `빅데이터(Row & Column)` 또는 `알고리즘 복잡하게` 또는 `최적화`를 통해 해결
>
> - Test의 Variance가 높다면, `빅데이터(Row) & 스몰데이터(Column)` 또는 `알고리즘 덜 복잡하게` 또는 `최적화`를 통해 해결

> <center><img src='Image/Expert/Bias_Variance_Reduce.png' width='600'></center>
>
> - **딥러닝(인공지능 알고리즘):** 딥러닝은 `엄청나게 복잡한 모델`이며 `Bias-variance Trade-off를 피할 수 없음`
>
> - `스몰데이터`의 딥러닝은 `과대적합`되어 `High Variance`가 우려되기에, `딥러닝`으로 성능을 내기 위해선 `빅데이터`가 반드시 필요!
>
> - `빅데이터`를 통해 `Train과 Test의 패턴 차이 감소`되어 `Bias & Variance를 모두 감소시키기 유리`

## Ensemble 알고리즘 종류

> **"`데이터 분리/결합` 및 `알고리즘 분리/결합`에 따라 다양하게 존재"**
>
> - **보팅(Voting):** `여러개의 알고리즘`의 결과들 중 `투표`를 통해 최종 `예측`
>
> - **배깅(Bagging):** `같은 알고리즘`으로 `여러개로 나눈 데이터 샘플`로 학습한 후 `보팅` 방식으로 `예측` 
>
> - **부스팅(Boosting):** `여러개의 알고리즘`이 `순차적`으로 학습하되, 앞선 알고리즘의 `오차/에러`에 `가중치`를 부여하여 `학습튜닝 및 예측 향상`
>
> - **스태킹(Stacking):** `여러개의 알고리즘` 결과를 `새로운 학습 데이터`로 만들어 다른 알고리즘에 `재학습 및 예측`
>

**1) 보팅(Voting):** `여러개의 알고리즘`의 결과들 중 `투표`를 통해 최종 `예측`

- **Hard Voting:** 다수 알고리즘 예측 결과값들을 `다수결`을 통해 `최종 레이블` 선정
- **Soft Voting:** 다수 알고리즘 예측 결과값들의 `확률을 평균`하여 `최종 레이블` 선정

> - 일반적으로 `Soft Voting`의 성능이 우수한 편

**2) 배깅(Bootstrap Aggregating, Bagging):** `같은 알고리즘`으로 `여러개로 나눈 데이터 샘플`로 학습한 후 `보팅` 방식으로 `예측` 

- **Voting:** `같은 데이터셋` 기반으로 `여러개의 알고리즘`의 결과들 중 `투표`를 통해 최종 `예측`
- **Bagging:** `여러개로 나눈 데이터 샘플` 기반으로 `같은 알고리즘`의 결과들 중 `보팅`을 통해 최종 `예측` 

> - 전체 데이터에서 샘플 분류시 `중첩을 허용`하는 방식을 `BootStraping` 분할 방식

- **불순도 측정 지표:** `엔트로피(Entropy)` & `지니(Gini Index)`

> **(1) 엔트로피(Entropy):** 분포의 순수도(Purity) 나타내는 척도로 순도가 낮으면 `엔트로피 높고` 순도가 높으면 `엔트로피 낮음`, `무질서도`라고도 함
>
> <center><img src='Image/Expert/Ensemble_EntropyExample.png' width='600'>(https://tyami.github.io/machine%20learning/decision-tree-1-concept/)</center>
>
> <br>
>
> - **확률 $\rightarrow$ 정보량 $\rightarrow$ 엔트로피 $\rightarrow$ 정보획득** 
>
>
> - **확률(Probability):** `예상하지 못한 사건` 발생 확률
>
> - **정보량(Information):** `잘 일어나지 않는 사건`은 자주 발생하는 사건보다 `정보량 많음` 가정
>
> $$
\begin{aligned}
\text{예상못한 사건 발생 확률(Probability)} &= p(x) \\
\text{예상못한 사건으로 알게될 정보(Information)} &= I(x) = \log_2 \frac{1}{p(x)} \\
\end{aligned}
$$
>
> <center><img src='Image/Expert/Ensemble_Information.png' width='600'>(https://tyami.github.io/machine%20learning/decision-tree-2-ID3/)</center>
>
> <br>
>
> - **엔트로피(Entropy):** `어떤 사건`이 나올지 `예측하기 어려운 정도`
>
> $$
\begin{aligned}
\text{예상못한 사건으로 알게될 정보(Information)} &= I(x) = \log_2 \frac{1}{p(x)} \\
\text{예상하지 못한 평균 정보(Entropy)} &= H(S) \\
&= \sum_{i=1}^N p_i\log_2 \frac{1}{p_i}=-\sum_{i=1}^N p_i\log_2 p_i
\end{aligned}
$$
>
> - **정보획득(Information Gain):** `분류 전후`의 `변화된 정보량`을 통해 `분류성능`을 확인하며, 분류가 잘 되지 않아 별 차이가 없으면 `낮은 수치`, 잘 분류되어 변화가 많으면 `높은 수치`
>
> $$
\begin{aligned}
\text{Information Gain} &= \text{Base Entropy} - \text{New Entropy} \\
&= \text{H(S)} - \text{H(S | Each Case)} \\
\end{aligned}
$$
>
> **(2) 지니(Gini Index):** 불확실성(묶여있는정도)을 수치화 한 것으로, 여러 레이블들이 묶이면 `수치가 높고` 유사 레이블들끼리 묶이면 `수치가 낮음`
> - `경제 불평등 지수`로 처음 등장을 하였으며, `소수에게 집중되면` 수치가 높고 `개개인에게 모두 분배`되면 수치가 낮음

### ID3 알고리즘

\end{aligned}

### C4.5 / C5.0 알고리즘

> **"ID3 알고리즘 대비 기능적인 변경 및 추가"**

**1) 비용함수 변경:** 비용함수로 `Entropy`를 사용하고, 성능 비교를 위해 `Information Gain Ratio` 사용

- **정보획득(Information Gain) 한계:** 가지 갯수가 많을수록 높은 정보획득이 나오기 때문에 `높은 나무`가 되는 것이 아니고 가지가 많은 `잡초` 되는 격

<center><img src='Image/Expert/DecisionTree_IGLimit.png' width='400'></center>

$$
\begin{aligned}
\text{Information Gain} &= H(7,3) - (\frac{6}{10}H(6,0)+\frac{4}{10}H(1,3)) \\ 
&=0.8813-(\frac{6}{10} \times 0+\frac{4}{10} \times 0.8113) \\ 
&=0.5568
\end{aligned}
$$

<center><img src='Image/Expert/DecisionTree_IGLimit2.png' width='600'>(https://tyami.github.io/machine%20learning/decision-tree-3-c4_5/)</center>

$$
\begin{aligned}
\text{Information Gain} &= H(7,3) - (\frac{1}{10}H(1,0)+\frac{1}{10}H(1,0)+…+\frac{1}{10}H(0,1)+\frac{2}{10}H(1,1)) \\ 
&=0.8813-(\frac{1}{10} \times 0+…+\frac{2}{10} \times 1) \\ 
&=0.6813
\end{aligned}
$$

- **정보획득비율(Information Gain Ratio, GR):** 가지의 갯수가 많을수록 높은 정보획득이 나오는 `한계를 보완`하고자 `무분별한 노드생성에 패널티`를 주는 정교한 불순도 측정지표

$$
\begin{aligned}
\text{Information Gain Ratio} &= \frac{\text{Information Gain}}{\text{Base Entropy}} \\
\end{aligned}
$$

> - **Example1:**
> 
> $$
\begin{aligned}
\text{Information Gain Ratio}&=\frac{0.5568}{-(\frac{6}{10}log_2 \frac{6}{10} + \frac{4}{10}log_2 \frac{4}{10})} \\ 
&=\frac{0.5568}{0.9701} \\ 
&=0.5739
\end{aligned}
$$
> - **Example2:**
> 
> $$
\begin{aligned}
\text{Information Gain Ratio}&=IG/IV \\ 
&=\frac{0.6813}{3.1219} \\ 
&=0.2182
\end{aligned}
$$

---

**2) 연속형 데이터 반영:** `범주형 변수` 뿐 아니라 `연속형 변수`도 분류값으로 반영하여 추정

**3) 결측치 처리:** Information Gain을 `비결측치` 데이터의 `Weighted` 비율로 반영하여 해결

**4) 과적합 방지:** 

- 의사결정나무는 특정 기준이 없어질 때까지 `무한히 커질 가능성` 존재
- `너무 큰 나무`처럼 깊어질수록, Train은 잘 분류하지만 `Test`는 잘 분류하지 못하는 `과적합 발생`
- 나무가 `너무 깊어지지 않도록` 중간 노드들을 잘르는 `가지치기(Pruning)` 과정 필요
- `인공신경망` 알고리즘의 `Dropout`의 아이디어가 됨

> - **사전 가지치기(Pre-pruning):** 의사결정나무를 `완성하기 전`에 가지치기 수행
>> - `특정 갯수의 깊이`를 정해두고 그 수에 도달하면 나무의 성장을 멈춤
>> - `Node / Leaf`에 속하는 `데이터의 최소 갯수`를 정해두고 그 수에 못미치면 나무 성장을 종료
>> - 어느 정도 깊이가 좋은지 미리 알 수 없고 중요 패턴을 놓칠수도 있음
>> - `인공신경망` 알고리즘의 `Early Stopping`의 아이디어 유사
>
> - **사후 가지치기(Post-pruning):** 의사결정나무를 `완성한 후`에 가지치기 수행
>> - `의도적`으로 나무를 `크게 성장`시킨 후, 적절한 높이로 가지치기
>> - 최적의 나무 크기를 알 수 없기 때문에 `사전 가지치기 보다 효율적` 방법
>> - `인공신경망` 알고리즘의 `Dropout`과 차이는 있지만 방향 및 아이디어 유사
>
> <center><img src='Image/Expert/DecisionTree_Pruning.jpg' width='900'>(가지치기 종류: https://dev-hunmin.tistory.com/)</center>

- `C4.5` 알고리즘은 `사전 가지치기` 방식 채택

> (1) `Train/Test` 대신에 `Train/Pruning/Test` 데이터로 분류
>
> (2) `Train` 데이터로 `의사결정나무 생성`
>
> (3) `Pruning` 데이터로 자식 노드부터 부모 노드까지의 `오류/에러`를 계산하여, 자식노드 보다 `부모노드의 에러가 작을 경우` 해당 분기 `가지치기`
>
> (4) 완성된 `Pruned Decision Tree` 사용하여 Test 데이터 예측

### CART 알고리즘

> **"ID3 알고리즘과 비슷한 시기에 개발된 `분류/예측` 알고리즘"**

**1) 비용함수:** 비용함수로 `Entropy` 대신 `Gini Index` 사용

- 엔트로피처럼 `분류가 잘` 될 때 `낮은 값`이 제시되고, 가장 낮은 지표 찾아 분기

$$
\begin{aligned} 
G(S) &= \sum_{i=1}^{N} p_i(1-p_i) \\
&= \sum_{i=1}^N pi - \sum_{i=1}^N pi^2 \\
&= 1 - \sum_{i=1}^N p_i^2 \\
\end{aligned}
$$

<center><img src='Image/Expert/Entropy_Gini.png' width='900'>(https://quantdare.com/decision-trees-gini-vs-entropy/)</center>

---

**2) Binary Tree:** 

> - 여러개의 자식노드로 분기되지 않고, 오로지 `2개의 노드로만 분기`
> - 선택된 변수에서만 분기하는 ID3과 달리 변수 내 모든 레이블만큼 분기하지 때문에, `변수 내 각 레이블 마다의 IG 계산`되어 `하나의 레이블과 나머지`로 가지가 생성되어 훨씬 많은 연산 필요
>> - ID3 $$
IG(Play,Outlook) \\ 
IG(Play,Temeperature) \\ 
IG(Play,Humidity) \\ 
IG(Play,Windy)
$$
>> - CART $$
IG(Play,Outlook=sunny) \\ 
IG(Play,Outlook=overcast) \\ 
IG(Play,Outlook=rain) \\ 
IG(Play,Temperature=hot) \\ 
IG(Play,Temperature=mild) \\ 
IG(Play,Temperature=cool) \\ 
IG(Play,Humidity=high) \\ 
IG(Play,Humidity=normal) \\ 
IG(Play,Windy=TRUE) \\ 
IG(Play,Windy=FALSE) \\
$$
> 
> <center><img src='Image/Expert/DecisionTree_CART_BinaryTree.png' width='900'>(https://tyami.github.io/machine%20learning/decision-tree-4-CART/)</center>

**3) Regression Tree:**

> - 성별, 등급과 같이 레이블이 아니라 온도, 가격 등의 `값을 예측하는 문제`에도 사용 가능
> - 비용함수로 `불순도 지표`가 아닌, 실제값과 예측값의 `오차`를 사용
> - 특정 값을 기준으로 `구간을 나누어` 의사결정나무로 분류하고 새로운 `값이 속한 구간의 평균값`으로 예측
>
> <center><img src='Image/Expert/DecisionTree_CART_Regression.png' width='900'></center>
>
> <center><img src='Image/Expert/DecisionTree_CART_RegressionLable.png' width='900'></center>
>
> - **비용함수:** `좌우가지`에서의 `오차값`과 `데이터 수의 비율`을 가중치로 곱해 분기 결정
>
> $$
J(m, MSE) = \frac{m_{left}}{m}MSE_{left}+\frac{m_{right}}{m}MSE_{right}
$$

**4) 과적합 문제:**

> - 나무가 `깊어질수록` 촘촘하게 구간이 생기고 `오차는 줄어들 것`
> - 만약 `Train 모든 수치값`에 맞춰 의사결정나무가 `생성`되면 Train에 너무 `과하게 학습`되어 과적합 유발
>
> <center><img src='Image/Expert/DecisionTree_Overfitting.png' width='900'>(https://tyami.github.io/machine%20learning/decision-tree-4-CART/)</center>

> - `분류/예측` 모든 경우에 과적합 이슈가 있으며, `가지치기`로 줄일 수 있더라도 `한계` 존재
>> - 나무의 최대 깊이는?
>> - 자식 노드에 포함될 최소 샘플 수는?
>> - 불순도/오차의 최소값 기준은?
>

````python
# 필요 라이브러리 호출
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import numpy as np
import scipy as sp
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib import gridspec
from sklearn import preprocessing
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.tree import DecisionTreeRegressor, DecisionTreeClassifier
from sklearn.tree import plot_tree, export_text, export_graphviz
from mlxtend.plotting import plot_decision_regions

# 예제 데이터 로딩
rng = np.random.RandomState(123)
X_train = np.sort(5 * rng.rand(500, 1), axis=0)
Y_train = np.sin(X_train).ravel()                        # sin함수의 예측을 목표로한다
X_test = np.arange(0.0, 5.0, 5/X_train.shape[0])[:, np.newaxis]
Y_test = Y_train.copy()
Y_test[::5] = Y_test[::5] + 3 * (0.5 - rng.rand(100))

# 예제 데이터 시각화
plt.figure(figsize=(10,10))
plt.scatter(X_test, Y_test, s=20, c='darkorange')
plt.title('Example Data')
plt.xlabel('X')
plt.ylabel('Y')
plt.axis("equal")
plt.grid()
plt.show()
````

````python
# 모델링: Decision Tree
## 학습
model_dt1 = DecisionTreeRegressor(criterion='squared_error', max_depth=3,
                                  random_state=123)
model_dt2 = DecisionTreeRegressor(criterion='squared_error', max_depth=5,
                                  random_state=123)
model_dt1.fit(X_train,Y_train)
model_dt2.fit(X_train,Y_train)
## 학습트리 시각화 및 출력
## 같은 색 계열이면 같은 집단으로 분류를 한 것
## 색이 진할수록 비용함수가 낮은 것
## 마지막 자식 노드에서 비용함수가 0에 가까운 것은 잘 분류? 또는 과대적합?
plt.figure(figsize=(15,15))
plot_tree(model_dt1, impurity=True, filled=True, rounded=True)
plt.show()
print(export_text(model_dt1))
plt.figure(figsize=(15,15))
plot_tree(model_dt2, impurity=True, filled=True, rounded=True)
plt.show()
print(export_text(model_dt2))
## 예측
Y_tepred1 = model_dt1.predict(X_test)
Y_tepred2 = model_dt2.predict(X_test)
## 학습결과
plt.figure(figsize=(10,10))
plt.scatter(X_test, Y_test, s=20, c='darkorange')
plt.plot(X_test, Y_tepred1, color='cornflowerblue', linewidth=4, label='max_depth=3')
plt.plot(X_test, Y_tepred2, color='yellowgreen', linewidth=2, label='max_depth=5')
plt.title('Decision Tree Regression')
plt.xlabel('X')
plt.ylabel('Y')
plt.axis("equal")
plt.grid()
plt.legend()
plt.show()
````

### 사용방법 실습

````python
# 필요 라이브러리 호출
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import numpy as np
import scipy as sp
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib import gridspec
from sklearn import preprocessing
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.tree import DecisionTreeRegressor, DecisionTreeClassifier
from sklearn.tree import plot_tree, export_text, export_graphviz
from mlxtend.plotting import plot_decision_regions
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier

# 예제 데이터 로딩
np.random.seed(0)
rv0 = sp.stats.multivariate_normal([-2, -2], [[2, 0.9], [0.9, 2]], seed=123)
rv1 = sp.stats.multivariate_normal([1, 1], [[2, -0.8], [-0.8, 2]], seed=123)
X0 = rv0.rvs(100)
X1 = rv1.rvs(30)
X = np.vstack([X0, X1])
Y = np.hstack([np.zeros(100), np.ones(30)])

# 예제 데이터 시각화
plt.figure(figsize=(10,10))
xx1 = np.linspace(-5, 5, 100)
xx2 = np.linspace(-5, 5, 100)
XX1, XX2 = np.meshgrid(xx1, xx2)
plt.contour(XX1, XX2, rv0.pdf(np.dstack([XX1, XX2])), cmap=mpl.cm.cool)
plt.contour(XX1, XX2, rv1.pdf(np.dstack([XX1, XX2])), cmap=mpl.cm.hot)
plt.scatter(X0[:, 0], X0[:, 1], marker="o", c='b', label="y=0")
plt.scatter(X1[:, 0], X1[:, 1], marker="s", c='r', label="y=1")
plt.legend()
plt.title('Example Data')
plt.axis("equal")
plt.grid()
plt.show()

# 데이터 분리 및 스케일링
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=123)
scaler = preprocessing.StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
````

````python
# 모델링: Bagging
## 학습
model_rf = RandomForestClassifier(n_estimators=100, random_state=123)
model_rf.fit(X_train, Y_train)
## 학습트리 시각화 및 출력
## 같은 색 계열이면 같은 집단으로 분류를 한 것
## 색이 진할수록 비용함수가 낮은 것
## 마지막 자식 노드에서 비용함수가 0에 가까운 것은 잘 분류? 또는 과대적합?
for index in range(0,5):
    plt.figure(figsize=(15,15))
    plot_tree(model_rf[index], impurity=True, filled=True, rounded=True)
    plt.show()
    print(export_text(model_rf[index]))
## 학습결과
plt.figure(figsize=(10,8))
plot_decision_regions(X=X_train, y=Y_train.astype(np.int_), clf=model_rf)
plt.title('Random Forest')
plt.show()
## 예측
Y_trpred = model_rf.predict(X_train)
Y_tepred = model_rf.predict(X_test)
````

## **Boosting:** Bias 줄이기 위한 목적

**"여러개의 알고리즘이 `순차적으로 학습`하되, `첫 알고리즘`이 `전체 데이터`를 학습하고 앞선 알고리즘에서 잘 학습되지 않은 `오차/에러`에 중요도를 높여 `가중치`를 부여하고 반복적으로 `모델링 업데이트`를 위한 학습 및 예측"**

- **장점:** 

> - Bagging과 달리 `맞추기 어려운 문제`를 `학습`하는데 특화
> - 학습되지 않은 패턴을 계속 반영하기 `Bias를 줄이는데 효과적`
> - `Outlier`나 `Anomaly Detection`에 강함

- **단점:** 

> - `잘 학습된 특징`들은 중요도를 낮춰 `덜 고려함`
> - 학습하지 못한 데이터 패턴들을 학습하기 때문에 `직렬처리로 느린 편`

### Adaptive Boosting(AdaBoost) 알고리즘

**1) Random Forest 비교:** 

> **(1)** Random Forest의 Decision Tree 대신 `Stump(Binary Tree)`를 사용
>
> **(2)** Random Forest에선 개별 알고리즘에 동일한 가중치를 사용했지만, AdaBoost는 `Stump 마다 서로 다른 가중치`를 부여하여 결과값이 달라질 수 있음
>
> **(3)** Random Forest는 Tree 들을 병렬로(Parallel) 한번에 생성하고 Tree 들 간 서로 독립적이지만, AdaBoost는 `기존 Stump가 다음 Stump에 영향`을 주며 `순서대로(Sequential) 생성` 
>
> <center><img src='Image/Expert/AdaBoost_Structure.png' width='700'></center>

---

**2) 알고리즘 추정방향:**

<center><img src='Image/Expert/AdaBoost_ExampleData.png' width='350'>(StatQuest)</center>

> **(1) First Stump:**
>
> - 몸무게 기준의 `Gini Index`가 가장 작기 때문에 `몸무게 Stump`를 첫번때 Stump로 지정
> - 몸무게 오름 차순 $\rightarrow$ 인접한 몸무게 평균 $\rightarrow$ 각 평균마다 불순도 계산 $\rightarrow$ 가장 작은 Gini Index 갖는 몸무게 평균값이 `176`
>
> <center><img src='Image/Expert/AdaBoost_ExFirst.png' width='600'></center>

> **(2) Amount of Say:** 최종 `분류 모델 예측값`에 해당 Stump가 `얼마만큼의 영향`을 주는지 `가중치`로 사용
>
> <br>
> <center><img src='Image/Expert/AdaBoost_ExFirstError.png' width='600'></center>
>
> $$
\begin{aligned} 
\text{Amount of say} &=\alpha_t=\frac{1}{2}log(\frac{1-\epsilon_t}{\epsilon_t}) \\
& \text{where } \epsilon_t \text{ is the total sum of misclassified sample weight}
\end{aligned} 
$$
>
> - `X축`은 Total Error, `Y축`은 Amount of Say
> - `Total Error가 0`이면 Amount of Say는 `큰 양수`이고, `Total Error가 1`이면 Amount of Say는 `큰 음수`
> - `Total Error가 0`이면 `올바른 분류`를 한 알고리즘이고, `1`이면 `반대로 분류`를 한 알고리즘
> - `Total Error가 0.5`일 때는 `Amount of Say = 0`, 동전을 던지는 것처럼 의미가 없다는 뜻
>
> <center><img src='Image/Expert/AdaBoost_ExAS.png' width='200'></center>
>
>  $$
\begin{aligned} 
\text{Amount of say} &=\alpha_t=\frac{1}{2}log(\frac{1- 1/8}{1/8}) \\
&= 0.97
\end{aligned} 
$$

> **(3) Updated Sample Weight:** 
>
> - 하나의 Stump가 `잘못 분류한 샘플`에 대해 다음 Stump로 넘겨줄 때 `가중치를 높여 전달`
>
> <br>
> <center><img src='Image/Expert/AdaBoost_ExUpdatedSample1.png' width='600'></center>
>
> - `잘 분류한 샘플`에 대해선 `가중치를 낮춰 전달`
> - 전체 가중치의 합을 1로 만들기 위해 `정규화`
>
> <br>
> <center><img src='Image/Expert/AdaBoost_ExUpdatedSample2.png' width='500'></center>

> **(4) Next Stump:** 
>
> **(4-1):** Resampling
> - 최종 Sample Weight를 기준으로 `index`를 `0~1` 사이의 구간으로 구분
> - `원데이터의 샘플 수`만큼 `0부터 1`사이의 숫자를 `Random Sampling` 하고, 이 값에 따라 선택된 `index 샘플`로 원데이터를 새롭게 구성
> - `Sample Weight가 컷던 샘플`은 더 넓은 구간값이기 때문에, `중복되어 더 자주 추출`
> - `Sample Weight`는 모든 샘플이 동일하게 `초기화`
>> - Sample Weight는 동일하지만, 잘못 분류되었던 샘플 비율이 더 높기 때문에 결과적으로 오분류 샘플의 가중치가 다른 것보다 높은 효과
>
> <br>
> <center><img src='Image/Expert/AdaBoost_ExNextStump.png' width='600'></center>
>
> **(4-2):** `(1)`적용 및 재추정하여 `Next Stump` 생성
>
> <center><img src='Image/Expert/AdaBoost_ExFirst.png' width='600'></center>

> **(5) Repetition:** 정해진 반복만큼의 Stump 생성시 까지 `(1) + (2) + (3) + (4)`를 반복하여 `모든 Stump 생성`

> **(6) Final Decision:** 정해진 반복만큼의 Stump 생성시 까지 `(1) + (2) + (3) + (4)`를 반복하여 `모든 Stump 생성`
>
> - `각 Stump`마다의 `Amount of Say` 추정
> - 최종적으로 계산된 `Amount of Say` 평균값으로 `Heart Disease`가 있다고 분류 예측
> - 모든 Stump의 예측값 $h_t(x)$에 Amount of Say $\alpha_t$를 가중치로 곱하는 회귀함수 방식인 `Hard Voting`으로 최종 예측값 선택
>>
>> $$
F(x)=\sum_{t=1}^N \alpha_t h_t(x)
$$
>

### Gradient Boosting Machine(GBM) 알고리즘

> **"AdaBoost 알고리즘 대비 `기능적인 변경 및 추가`"**

---

**1) 개별 알고리즘(Weak Learner) 변경:**

- AdaBoost에선 Stump를 사용했지만, GBM에선 `Restricted Tree` 사용
- `First Weak Learner`는 모든 데이터의 평균값을 갖는 하나의 초기 추정값인 `Single Leaf`에서 출발
- `나머지 Restricted Tree`는 `최대 잎의 수`를 제한을 걸어 과적합 제한

<center><img src='Image/Expert/AdaBoost_GBM.png' width='700'></center> 

---

**2) 가중치 방식 변경:** AdaBoost는 각 Weak Learner의 가중치(크기)가 다른 반면, `GBM`에선 `동일 가중치`

> **"AdaBoost Equation"**
>
> $$
F(x)=\sum_{t=1}^N \alpha_t h_t(x)
$$
>
> **"GBM Equation"**
>
> - AdaBoost는 Amount of Say를 가중치로 사용한 반면, `GBM`에선 모두 동일한 `Learning Rate` $\eta$ 사용
>
> $$
F(x)=F_0(x) + \eta \sum_{t=1}^N h_t(x)
$$


---

**3) 예측값 변경:** Tree 예측값인 $h_t(x)$가 다르기 때문

> - AdaBoost Stump들은 모두 실제 회귀/분류 값을 예측하기 때문에, 이 값들을 가중 평균하여 실제값 예측
> - `GBM`은 각 `Restricted Tree들의 예측값`은 실제 회귀/분류 값이 아닌 `이전 Tree의 오차/잔차(Pseudo Residual)`
> - `최종 예측값`은 각 모델의 `오차를 스케일링 및 합산`하여 실제값 예측값
>

### eXtreme Gradient Boosting(XGBoost) 알고리즘 

> **"GBM을 발전시킨 알고리즘이 XGBoost, Light GBM, CatBoost"**

- `XGBoost(2014년), LightGBM(2016년), CatBoost(2017년)`은 더 빠르고 성능 좋은 알고리즘으로, Kaggle에서 상위권 Ranker들이 많이 사용하는 알고리즘
- GBM의 `Overfitting 문제를 보완`하기 위해 개발
- `분류` 뿐만 아니라 `회귀`에서도 뛰어난 `예측 성능`
- `CPU, GPU, 병렬처리` 등을 지원하여 GBM 대비 `빠른 실행시간`
- `Regularization, Tree Pruninng, Cross-validation` 등 다양항 성능 향상 최적화 기능
- `max_depth`로 분할 깊이를 조정하지만, `Tree Pruning`으로 더 이상 이득 없는 분할 수를 더 줄이는 장점

> - Approximate Greedy Algorithm
> - Parallel learning
> - Weighted Quantile Sketch
> - Sparsity-Aware Split Finding
> - Cache-Aware Access
> - Blocks for Out-of-Core Computation

- Regularization을 포함한 `다양한 비용함수` 지원
- `Early Stopping, 자체 교차 검증, 결손값 처리` 등 다양한 편의 기능
- `튜닝`해야 할 `하이퍼파라미터가 너무 많아` 단점

### LightGBM 알고리즘

- GBM은 모든 데이터를 탐색/학습하기에 `학습 시간이 오래걸리는 문제`를 보완하기 위해 개발
- XGBoost의 장점은 계승하고 `단점은 보완`하는 방식으로 개발
- `병렬 컴퓨팅`을 지원하며 최근에는 CPU외에 `GPU`까지 사용 가능
- 메인 기술은 `GOSS(Gradient-based One-Side Sampling)`로 Information Gain을 가중치가 작은 부분의 데이터에 집중하기 때문에 `학습이 덜 된 데이터`에 초점
- 현존하는 부스팅 알고리즘 중 `가장 빠르고 높은 예측력` 제공(특히 속도)
- 스몰데이터(약10000건 정도)에서 `과적합` 우려 
- GBM은 항상 균형을 이루는 Binary 분할을 하는 `Level-wise(균형 트리 분할)` 방식을 사용하나, LightGBM은 `Leaf-wise(리프 중심 트리 분할)`을 사용하기 때문에 시간과 메모리 모두 절약 가능

## **Bagging vs Boosting**

<!-- <center><img src='Image/Expert/Bagging_Boosting2.png' width='600'></center>  -->

<center><img src='Image/Expert/Bagging_Boosting.png' width='300'></center> 

<center><img src='Image/Expert/Ensemble_FutureDirection.png' width='600'>(https://seokchaeyoon.github.io/)</center> 

|  	| **Bagging** 	| **Boosting** 	|
|:---:	|:---:	|:---:	|
| **설명** 	| Bootstrapping을 통해 여러 학습 데이터를 만들고,<br> 각 학습데이터의 예측치의 평균으로 최종 예측값 선정 	| 여러 알고리즘들에 데이터를 순차적으로 학습 및<br> 오차를 줄이며 최종 예측값 선정 	|
| **목적** 	| Bias가 낮은 알고리즘들을 사용하여<br> `예측 Variance를 줄이는 것` 	| Variance가 낮은 알고리즘들을 사용하여<br> `예측 Bias를 줄이는 것` 	|
| **특징** 	| `병렬` 앙상블 모델<br>(각 모델은 서로 독립) 	| `연속` 앙상블 모델<br>(이전 모델의 오류 반영) 	|
| **적합한 상황** 	| Low Bias + `High Variance` 	| `High Bias` + Low Variance 	|

---

> **"경험적으로 `Boosting`이 Bagging보다 성능이 좋은 경우가 많지만, 엄청난 컴퓨팅 학습시간과 파라미터 튜닝에 소요되는 시간과 비용을 제외한다면... `Bagging`이 가성비라고 볼 수도 있음"**
>
> - 책상에서 `테스팅`은 Bagging
> - 실제 비즈니스 현장에 `릴리즈`는 Boosting

## **Stacking:** Bagging 또는 Boosting과 유사

> **"`개별 알고리즘`을 서로 `결합하여 예측`"**

---

- 배깅 또는 부스팅과 유사하지만, 가장 큰 차이는 `개별 알고리즘`으로 `예측한 데이터`를 학습하여 `다시 예측`
- 개별 알고리즘의 예측 데이터를 `메타 데이터`라고 지칭하는 편
- 메타 데이터를 `별도의 알고리즘`으로 `최종 학습` 후 `최종 예측`하며 이때의 알고리즘을 `메타 알고리즘`라고 지칭하는 편
- 스태킹 모델링 시, `2종류 알고리즘` 필요

> (1) 개별적인 모델링에 사용할 `개별 알고리즘`
>
> (2) 개별 기반 예측 데이터를 학습하고 예측하는데 필요한 `메타 알고리즘`

<center><img src='Image/Expert/Stacking_Structure.png' width='500'>(http://rasbt.github.io/mlxtend/user_guide/classifier/StackingClassifier/)</center> 

- `캐글과 같은 대회`에서 높은 순위를 차지하기 위해 성능 수치를 높여야 할 경우 자주 사용됨
- 일반적으로 `성능이 비슷한 모델을 결합`해서 좀더 나은 `성능 향상`을 도출
- 2~3개의 개별 모델만을 결합해서 쉽게 예측 성능을 향상시킬 수 없음
- 스태킹을 적용한다고 해서 반드시 성능 향상 된다는 보장 없음
- `과적합` 단점을 개선하기 위해 다양한 `CV(Cross Validation)` 방식을 사용한, 다양한 `CV기반 Stacking` 알고리즘 존재
