# **Problem:** 기부액 향상을 위한 가장 효과적인 타겟 마케팅 AI모델 구축 전략

**1. 문제 설명:**

- **현상(Problem):** 현재 `우리의 마케팅 캠페인은 비용 대비 총기부액(ROI)이 지속적으로 정체되거나 하락`하고 있음

> **Why1:** 왜 기부 총액이 정체되고 있는가?
> - 전통적인 매스 마케팅(Mass Marketing) 방식으로 `불특정 다수에게 캠페인 예산을 균등하게 소진`하고 있기 때문

> **Why2:** 왜 불특정 다수에게 예산을 소진하는 것이 문제인가?
> - 기부 의사가 전혀 없는 사람에게 `마케팅 비용이 낭비`되고, 정작 `기부 잠재력이 높은 사람에게 충분한 리소스(맞춤형 메시지, 혜택 등)를 집중하지 못하기 때문`

> **Why3:** 왜 기부 잠재력이 높은 사람을 사전에 선별하지 못했는가?
> - 과거의 단순 통계나 직관만으로는, 고객의 정량적 데이터(과거 기부액, 소득 수준 등)와 정성적 데이터(웹사이트 체류 시간, 피드백 텍스트 등) 간의 `복잡하고 비선형적인 패턴을 인간이 모두 파악할 수 없기 때문`

> **Why4:** 왜 정량/정성적 패턴을 모두 학습하는 AI 모델이 필요한가?
> - AI 모델은 수많은 변수 속에서 `기부라는 행동(Action)을 유발하는 숨겨진 인과적 트리거(Trigger)를 찾아내고`, 각 개인의 기부 확률과 예상 기부액을 가장 정확하게 산출할 수 있기 때문

> **Why5:** 왜 이 AI 모델이 마케팅의 새로운 패러다임이며 기부액 향상으로 직결되는가?
> - AI가 도출한 `타겟 리스트를 바탕으로 예산을 최적화하여 재분배`한다면, 이탈 방지가 필요한 고액 기부자, 넛지(Nudge)가 필요한 잠재 기부자 등 `타겟별로 차별화된 마케팅 액션을 처방(Prescription)`함으로써, `동일한 예산으로도 전체 기부액을 극대화`하는 실질적인 임팩트를 창출

---

**2. 전처리 가이드라인:** 

- 우리는 `미래(Test) 데이터를 볼 수 없는데 전처리가 어떻게 가능`할까요?     
- 과제들에서 수업 내용과 무관한 비현실적 전처리를 진행하여 감점 + `AI 한테 제대로 알고 지시하지 못하면 AI는 마음대로 전처리` 진행
- 따라서 `평가의 형평성을 위해 아래와 같은 전처리 필수 통일 사항`을 요청 (불일치시 0점 처리)

| **순서** | **전처리 단계** | **방식** |
|:---:|:---:|---|
| **0** | **데이터 결합** | 절대 미허용 |
| **1** | **데이터 필터** | 샘플 절대 미허용 + 컬럼 자유롭게 허용 |
| **2** | **결측치 처리** | 삭제 미허용 + 그외 자유롭게 허용 |
| **3** | **이상치 처리** | 행(Row) 삭제 방식 절대 미허용 (특정값 대체나 Capping 등은 허용) |
| **4** | **데이터 분리** | - 함수: train_test_split<br>- Y 변수명: "기부여부" (1: 기부, 0: 미기부)<br>- 방식: 랜덤 분리<br>- Ratio: Test 20% + Train/Validation 자유롭게 진행<br>- Y 비율 보존: Stratify 반드시 사용<br>- 랜덤 시드: random_state=123 |
| **5** | **자료형 변환** | 자유롭게 허용 |
| **5** | **데이터 변환** | 자유롭게 허용 |
| **5** | **파생변수 생성** | 자유롭게 허용 |
| **5** | **샘플링** | - Train 데이터 한정 Under/Over 샘플링 자유롭게 허용<br>- Test 데이터 샘플링 절대 미허용 |
| **5** | **스케일링** | - Y 변환 절대 미허용<br>- X 스케일링 자유롭게 허용 |

> - 전처리 순서는 4번까지는 반드시 차례대로 진행하고 이후 단계들은 자유롭게 진행 및 미진행 여부나 순서 등 자유롭게 진행
> - 단, 이전 단계의 결과가 누적으로 다음 단계에 활용되어야 하며 별개의 과정으로 진행시 제대로된 전처리가 아니기에 0점 처리

**4. 평가:**

- 일반적으로 기부자(1)와 미기부자(0)의 비율은 `0이 훨씬 많은 불균형(Imbalance)` 상태 
- 따라서 무조건 `기부하지 않는다고 찍어도 정확도(Accuracy)는 높게 나올 수 있음`
- 타겟팅의 정확성과 잠재고객 확보를 동시에 달성하기 위해 `Test 데이터의 F1-Score를 최종 모델 성능 지표로 평가`
> - 정밀도(Precision): 모델이 "이 사람은 기부할 것이다"라고 예측한 사람 중 실제로 기부한 사람의 비율
> - 마케팅적 의미로 `예산 낭비 방지`이며, 정밀도가 낮으면 기부 안 할 사람에게 마케팅 비용(우편물, 사은품 등)을 낭비하게 됩니다.
> - 재현율(Recall): 실제 기부를 한 전체 사람 중에서 모델이 맞춘 사람의 비율
> - 마케팅적 의미로 `기회 손실 방지`이며, 재현율이 낮으면 기부할 의사가 충분히 있는 사람을 놓치게 됩니다.
> - F1-Score: Precision과 Recall의 조화 평균.
